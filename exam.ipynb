```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('img/tree.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def binary_function(pixel_value, trashold):
    return 1 if pixel_value <= thrÐµshold else 0
```

```python
binary_image = np.vectorize(binary_function)(image, 120)
```

```python
cv2.imshow('Original Image', image)
cv2.imshow('Binary Segmentation', binary_image.astype(np.uint8) * 255)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
intensity_values = np.arange(0, 256)

binary_values = np.vectorize(binary_function)(intensity_values, 120)

plt.plot(intensity_values, binary_values, label='Binary Characteristic Function')
plt.xlabel('Intensity')
plt.ylabel('Binary Value')
plt.title('Binary Characteristic Function')
plt.legend()
plt.show()
```
